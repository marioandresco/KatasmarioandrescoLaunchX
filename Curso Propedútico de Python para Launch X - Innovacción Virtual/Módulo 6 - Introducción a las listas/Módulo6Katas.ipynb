# Ejercicio 2: Trabajando con datos de una lista

```python
planets = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano', 'Neptuno']
```

```python
planeta_u = input('Por favor ingrese el nombre del planeta (con una letra mayúscula para comenzar)')
```

```python
planet_index = planets.index(planeta_u)
```

```python
print('Aquí están los planetas más cerca que ' + planeta_u)
print(planets[0:planet_index])
```

```python
print('Aquí están los planetas más lejos de ' + planeta_u)
print(planets[planet_index + 1:])
```
